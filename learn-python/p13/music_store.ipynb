```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# 读取数据集
music_data = pd.read_csv('datasets/music.csv')
music_data
```

```python
# 拆分出输出的数据集
X = music_data.drop(columns=['genre'])
X = X.values
X
```

```python
# 拆分出输出的数据集
y = music_data['genre']
```

```python
# 训练模型
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
# 预测21岁男性喜欢的音乐类型 和 22岁女性喜欢的音乐类型
predictions = model.predict([ 
    [21, 1], 
    [22, 0] 
])
predictions
```
